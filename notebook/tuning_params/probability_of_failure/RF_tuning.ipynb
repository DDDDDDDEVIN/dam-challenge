```python
train_X = pd.read_csv('../../../data/curated/probability_of_failure/train_X.csv')
train_Y = pd.read_csv('../../../data/curated/probability_of_failure/train_Y.csv')
val_X = pd.read_csv('../../../data/curated/probability_of_failure/val_X.csv')
val_Y = pd.read_csv('../../../data/curated/probability_of_failure/val_Y.csv')
test_X = pd.read_csv('../../../data/curated/probability_of_failure/test_X.csv')
test_Y = pd.read_csv('../../../data/curated/probability_of_failure/test_Y.csv')
```

```python
tunable_parameters = {
    'n_estimators': (25, 100, 400, 800),
    'max_depth': (2, 8, 32),
    'max_samples': (0.4, 0.7, 1),
    'max_features': (0.4, 0.7, 1),
    'ccp_alpha': (0.000001, 0.0001, 0.01, 0.1)
}
non_tunable_parameters = {
    'random_state':42
}
```

```python
from DevinMengTuner import DevinMengTuner
```

```python
model = RandomForestRegressor()
```

```python
tuner = DevinMengTuner()
tuner.set_data(train_X=train_X, train_Y=train_Y, test_X=val_X, test_Y = val_Y)
tuner.set_model(model, model_name='random_forest', model_type='Regression')
tuner.set_parameters(tunable_parameters=tunable_parameters, non_tunable_parameters=non_tunable_parameters)
tuner.tune()
```
