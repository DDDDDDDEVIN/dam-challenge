```python
import pandas as pd
from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import RandomForestRegressor
import sys
import warnings
from sklearn.decomposition import PCA
from sklearn.feature_selection import RFE
```

```python
warnings.filterwarnings('ignore')
```

```python
train_X = pd.read_csv('../../../data/curated/damage/train_X.csv')
train_Y = pd.read_csv('../../../data/curated/damage/train_Y.csv')
val_X = pd.read_csv('../../../data/curated/damage/val_X.csv')
val_Y = pd.read_csv('../../../data/curated/damage/val_Y.csv')
test_X = pd.read_csv('../../../data/curated/damage/test_X.csv')
test_Y = pd.read_csv('../../../data/curated/damage/test_Y.csv')
```

```python
estimator=RandomForestRegressor()
```

```python
rfe = RFE(estimator=estimator)
```

```python
rfe.fit(X=train_X, y=train_Y)
```

```python
rfe.support_
```

```python
train_X = train_X.iloc[:,rfe.support_]
val_X = val_X.iloc[:,rfe.support_]
test_X = test_X.iloc[:,rfe.support_]
```

```python
sys.path.append('../../../privatePackage')
```

```python
tunable_parameters = {
    'n_neighbors' : [i for i in range(1, 101)],
    'metric': ('manhattan', 'euclidean'),
    'weights': ('uniform', 'distance')
}
non_tunable_parameters = {
}
```

```python
from DevinMengTuner import DevinMengTuner
```

```python
model = KNeighborsRegressor()
```

```python
tuner = DevinMengTuner()
tuner.set_data(train_X=train_X, train_Y=train_Y, test_X=val_X, test_Y=val_Y)
tuner.set_model(model=model, model_name='KNN', model_type='Regression')
tuner.set_parameters(tunable_parameters=tunable_parameters, non_tunable_parameters=non_tunable_parameters)
tuner.tune()
```
