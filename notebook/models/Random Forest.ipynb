```python
import pandas as pd
from sklearn.metrics import r2_score, mean_squared_error
from sklearn.ensemble import RandomForestRegressor
import matplotlib.pyplot as plt
```

```python
train_X = pd.read_csv('../../data/curated/repair/train_X.csv')
train_Y = pd.read_csv('../../data/curated/repair/train_Y.csv')
val_X = pd.read_csv('../../data/curated/repair/val_X.csv')
val_Y = pd.read_csv('../../data/curated/repair/val_Y.csv')
test_X = pd.read_csv('../../data/curated/repair/test_X.csv')
test_Y = pd.read_csv('../../data/curated/repair/test_Y.csv')
```

```python
model = RandomForestRegressor(n_estimators=400, max_depth=32, max_samples=0.7, max_features=0.7, ccp_alpha=0.1)
```

```python
model.fit(train_X, train_Y)
```

```python
pred_Y = model.predict(test_X)
```

```python
r2_score(y_pred=pred_Y, y_true=test_Y), mean_squared_error(pred_Y, test_Y)
```

```python
importance_df = pd.DataFrame(data= [model.feature_importances_], columns=train_X.columns)
```

```python
plt.figure(figsize=(10, 8))
plt.bar(x=train_X.columns, height=model.feature_importances_)
plt.xticks(rotation=90)
```
